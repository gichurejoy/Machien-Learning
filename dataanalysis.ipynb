```python
from datetime import datetime
```

```python
df['Date'] = df.apply(lambda row: datetime(row['Year'], row['Month'], row['Day']).strftime('%d/%m/%Y'), axis=1)
```

```python
df.head()
```

```python
start_date = '01/01/2021'
end_date = '01/12/2021'
branch = 'NRB001'
item_code = None
filtered_data = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

if branch:
    filtered_data = filtered_data[filtered_data['Branch'] == branch]

if item_code:
    filtered_data = filtered_data[filtered_data['Item Code'] == item_code]
```

```python
grouped_data = df.groupby('Item Code')[['Total Qty IN', 'Total Qty OUT']].sum().reset_index()
print(grouped_data)
```

```python
import datetime

def calculate_oosd(purchases, sales):
    return max(purchases - sales, 0)

result = []
start_date = '01/01/2021'
end_date = '30/01/2021'

item_data = None
in_stock = False

for _, row in grouped_data.iterrows():
    total_purchases = row['Total Qty IN']
    total_sales = row['Total Qty OUT']

    actual_oosd = calculate_oosd(total_purchases, total_sales)


    if in_stock:

        formatted_end_date_oosd = end_date
    else:

        formatted_end_date_oosd = (datetime.datetime.strptime(start_date, '%d/%m/%Y') + datetime.timedelta(days=actual_oosd)).strftime('%d/%m/%Y')

    item_data = {
        "startDate": start_date,
        "endDate": formatted_end_date_oosd,
        "totalPurchases": total_purchases,
        "totalSales": total_sales,
        "actualOOSD": actual_oosd,
        "averageOOSD": actual_oosd
    }


    in_stock = total_purchases > total_sales

    result.append({
        "itemCode": row['Item Code'],
        "branch": "ALL",
        "itemData": item_data
    })

for item in result:
    print(item)
```
