```python
# Encoding:
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

Clustering_df['Age']=le.fit_transform(Clustering_df['Age'])
Clustering_df['Height']=le.fit_transform(Clustering_df['Height'])
Clustering_df['family_history_with_overweight']=le.fit_transform(Clustering_df['family_history_with_overweight'])
Clustering_df['SMOKE']=le.fit_transform(Clustering_df['SMOKE'])
Clustering_df['MTRANS ']=le.fit_transform(Clustering_df['MTRANS'])

Clustering_df.sample(5)
```

```python
data=Clustering_df.sample(frac=1).reset_index(drop=True)
data
```

## Clustering

Use k as either 3, 4 OR 5. (Only pick one of the three)

```python
# Use k - Means Clustering to cluster our data and assign each row to a cluster (assing labels to each row). 
# Be carefull with the arguments you pass to the function, make sure you pass the correct
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=4)
kmeans.fit(data)
print(kmeans.cluster_centers_)

# create a new column in our data and assign the cluster labels to each row: this has been partially done for you
print(kmeans.labels_)
data['New cluster'] = kmeans.labels_
data.head()
```

```python
# train a kNN model to predict the cluster of each row in our test data
# Be carefull with the arguments you pass to the function, make sure you pass the correct
from sklearn.neighbors import KNeighborsClassifier

# Create an instance of the KNeighborsClassifier class. Set the number of neighbors to 3

knn_clf=KNeighborsClassifier(n_neighbors=3)
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.25)

# Specify the columns to use for training i.e. Features (X) and the target column (y). Remeber the tagret is the cluster column from the clustering step

train_x =train.drop('New cluster', axis=1)
train_y=train['New cluster']

test_x =test.drop('New cluster', axis=1)
test_y=test['New cluster']

# Train / fit Features and Taregts to the KNeighborsClassifier modelknn_clf.fit(train_x,train_y)
knn_clf.fit(train_x,train_y)
predictions=knn_clf.predict(test_x)

from sklearn.metrics import accuracy_score
print(accuracy_score(test_y,predictions))

```
